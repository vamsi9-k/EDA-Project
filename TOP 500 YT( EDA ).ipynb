```python
def human_format(x):
    if x >= 1e9:
        return f'{x*1e-9:.0f}B'
    elif x >= 1e6:
        return f'{x*1e-6:.0f}M'
    elif x >= 1e3:
        return f'{x*1e-3:.0f}K'
    return str(int(x))
```

```python
x_labels = [human_format(x) for x in x_ticks]
```
