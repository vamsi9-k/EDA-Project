```python
import numpy as np
import pandas as pd 
import re 
import requests
from bs4  import BeautifulSoup
```

```python
# Requried URLs and Header for Data Collection using Web Scrapping
```

```python
url="https://us.youtubers.me/global/all/top-500-most-subscribed-youtube-channels"
request_header = {'Content-Type': 'text/html; charset=UTF-8','User-Agent': 'Chrome/101.0.0.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0','Accept-Encoding': 'gzip, deflate, br'}
```

```python
# Requesting to Access the url information
```

```python
page = requests.get(url,headers=request_header)
```

```python
page
```

```python
soup  = BeautifulSoup(page.text)
```

```python
soup
```

```python
# Data collection using Web Scrapping
```

```python
soup.find("div",class_="top-charts-box").find("a")
```

```python
URL=[]

for i  in soup.find_all("div",class_="top-charts-box"):
    for j in i.find_all("a"):
        url="https://us.youtubers.me/"+j.get("href")
        URL.append(url)
```

```python
URL
```

```python
Names=[]
Sub=[]
View=[]
Country=[]
Category=[]
Count=[]
Date=[]
Sub_view=[]
Sub_30=[]
Sub_90=[]
Est=[]
for i in URL:
    response=requests.get(i)
    soup=BeautifulSoup(response.text)
    #  For Loop code for Accesing container
    for i  in soup.find_all("div",class_="main col col-lg-8 col-md-8 col-sm-12"):
        for j in i.find_all("h1"):
            y=j.text
            # Extracting names
            names=re.findall(r"\b(.+)\sYoutube ",y)
            if names:
                Names.append(names[0])
            else:
                Names.append(np.nan)

                
     #  For Loop code for Accesing container
    for k in soup.find_all("div",class_="status-details"):
        t = re.sub(r"\n","",k.text.strip())
        # Extracting Subscribers
        sub=re.findall(r"^Subscribers\s+([\d,]+)",t)
        if sub:
            Sub.append(sub[0])
        else:
            Sub.append(np.nan)
    # Extracting Views
        view=re.findall(r"Subscribers\s+[\d,]+\s+Video views\s+([\d,]+)",t)
        if view:
            View.append(view[0])
        else:
            View.append(np.nan)
        # Extracting Count
        count=re.findall(r"Video count\s+(\d+)",t)
        if count:
            Count.append(count[0])
        else:
            Count.append(np.nan)
        # Extracting Country 
        country=re.findall(r"Country\s+([A-Za-z\s()]+)\s+Category",t)
        if country:
            Country.append(country[0])
        else:
            Country.append(np.nan)
        # Extracting Category
        cat=re.findall(r"Category\s+([\w]+)",t)
        if cat:
            Category.append(cat[0])
        else:
            Category.append(np.nan)
        # Extracting Youtuber Since 
        date=re.findall(r"Youtuber since \s+(\d{4})",t)
        if date:
            Date.append(date[0])
        else:
            Date.append(np.nan)

            
     #  For Loop code for Accesing container       
    for l in soup.find_all("div",class_="stats-board three-row"):
        t1=l.text.strip()
        re.sub(r"\n","   ",l.text.strip())
        sub_view=re.findall(r"([\d,]+)\s+last 7 days",t1)
        if sub_view:
            Sub_view.append(sub_view)
        else:
            Sub_view.append(np.nan)
        sub_view_30=re.findall(r"([\d,]+)\s+last 30 days",t1)
        if sub_view_30:
            Sub_30.append(sub_view_30)
        else:
            Sub_30.append(np.nan)
        sub_view_90=re.findall(r"([\d,]+)\s+last 90 days",t1)
        if sub_view_90:
            Sub_90.append(sub_view_90)
        else:
            Sub_90.append(np.nan)
            
    #  For Loop code for Accesing container
    for n in soup.find_all("div",class_="table"):
        t2=re.sub(r"\n","    ",n.text.strip())
        est=re.findall(r"Estimated earnings in past 30 days\s+(\$ [\d\.KM]+ - \$ [\d\.KM]+)",t2)
        if est:
            Est.append(est[0])
        else:
            Est.append(np.nan)
```

```python
len(Names)
```

```python
dic={
    "Youtuber":Names,"Subscribers":Sub,"Video views":View,"Total Video count":Count,"Country":Country,"Category":Category,"Youtuber since":Date,"Last 7 Days Subscribers":Sub_view,
     "Last 30 Days Subscribers":Sub_30,"Last 90 Days Subscribers":Sub_90,"Estimated earnings in past 30 days":Est
}
```
